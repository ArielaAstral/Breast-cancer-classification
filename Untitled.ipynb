```python
import tensorflow as tf
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
```

```python
Image_size=224
Batch_size=32
Channels=3
Epochs=200
```

```python
datasets=tf.keras.preprocessing.image_dataset_from_directory(
    "cancer",
    shuffle=True,
    image_size=(460,700),
    batch_size=Batch_size
)
```

```python
class_names=datasets.class_names
class_names
```

```python
import numpy as np
class1_count= 0

# Iterate over the dataset and count occurrences of each class
for size,labels in datasets:
    class1_count+= np.sum(labels.numpy())
class1_count
class0_count=2083-class1_count
class1_count,class0_count
```

```python
# Compute class weights
total_samples = 2083
class_weights = {
    0: (total_samples / 2) / class0_count,
    1: (total_samples / 2) / class1_count
}

class_weights
```

```python
def get_datasets_partition_tf(ds,train_split=0.7,val_split=0.15,test_split=0.15,shuffle=True,shuffle_size=10000):
    ds_size=len(ds)
    if shuffle:
        ds=ds.shuffle(shuffle_size,seed=12)
        
    train_size=int(train_split*ds_size)
    val_size=int(val_split*ds_size)
    
    train_ds=ds.take(train_size)
    val_ds=ds.skip(train_size).take(val_size)
    test_ds=ds.skip(train_size).skip(val_size)    
    
    return train_ds,val_ds,test_ds
```

```python
train_ds,val_ds,test_ds=get_datasets_partition_tf(datasets)
```

```python
train_ds=train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds=val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds=test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize_n_rescale=tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(Image_size,Image_size),
    layers.experimental.preprocessing.Rescaling(1.0/255)
])
```

```python
data_augmentation=tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
    layers.experimental.preprocessing.RandomRotation(0.2)
])
```

```python
Input_shape=(Batch_size,Image_size,Image_size,Channels)
n_classes=2
model=models.Sequential([
    resize_n_rescale,
    data_augmentation,
    
    layers.Conv2D(32,(3,3),activation='relu',input_shape=Input_shape),
    layers.MaxPooling2D(pool_size=(2,2)),
    
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64,activation="relu"),
    layers.Dense(2,activation="sigmoid"),
    
])
model.build(input_shape=Input_shape)
```

```python
import visualkeras
from PIL import ImageFont
font = ImageFont.truetype("arial.ttf")
visualkeras.layered_view(model, legend=True, font=font) # selected font
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
history=model.fit(
    train_ds,
    epochs=Epochs,
    batch_size=Batch_size,
    verbose=1,
    validation_data=val_ds,
    class_weight=class_weights
)
```

```python
score=model.evaluate(test_ds)
```
